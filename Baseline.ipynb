```python
import pandas as pd
import numpy as np
```

### Считывание данных

```python
train_df = pd.read_csv("train.csv")
```

```python
pd.set_option("max_info_columns", 500)
```

```python
train_df.info()
```

```python
train_df.columns
```

### Базовое решение

Нахождение среднего значения аудиосигнала

```python
mean_values = train_df[train_df.columns[:-1]].mean(axis=1, skipna=True)
X = np.reshape(np.array(mean_values), (-1, 1))
```

```python
y = train_df["label"]
```

Обучение модели дерева решения

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
model = model.fit(X, y)
```

### Вывод результата

```python
test_df = pd.read_csv("test.csv")
```

```python
test_x = np.reshape(np.array(test_df.mean(axis=1, skipna=True)), (-1, 1))
```

```python
predicted_label = model.predict(test_x)
```

```python
result_df = pd.DataFrame({"label":predicted_label})
result_df.to_csv("baseline_submission.csv", index=False)
```
